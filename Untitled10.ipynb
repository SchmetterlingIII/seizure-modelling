```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from IPython.display import Image
from google.colab import files

# Lorenz system
def lorenz(state, sigma=10, rho=28, beta=8/3):
    x, y, z = state
    return np.array([
        sigma * (y - x),
        x * (rho - z) - y,
        x * y - beta * z
    ])

# Simulate
dt = 0.01
state = np.array([1.0, 1.0, 1.0])
trajectory = [state]

for _ in range(1000):
    state = state + lorenz(state) * dt
    trajectory.append(state.copy())

trajectory = np.array(trajectory)

# Create animation
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

line, = ax.plot([], [], [], 'b-', alpha=0.6, lw=1)
point, = ax.plot([], [], [], 'ro', markersize=8)

ax.set_xlim(-20, 20)
ax.set_ylim(-30, 30)
ax.set_zlim(0, 50)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lorenz Attractor')

def update(frame):
    idx = frame * 10
    line.set_data(trajectory[:idx, 0], trajectory[:idx, 1])
    line.set_3d_properties(trajectory[:idx, 2])
    point.set_data([trajectory[idx, 0]], [trajectory[idx, 1]])
    point.set_3d_properties([trajectory[idx, 2]])
    return line, point

anim = FuncAnimation(fig, update, frames=100, interval=50, blit=True)

# Save as GIF
print("Saving GIF...")
anim.save('lorenz.gif', writer=PillowWriter(fps=20))
plt.close()

# Display in notebook
display(Image(filename='lorenz.gif'))

# Download the file to your computer
print("Downloading GIF...")
files.download('lorenz.gif')
```
